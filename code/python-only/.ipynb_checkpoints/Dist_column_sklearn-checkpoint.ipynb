# Minimal solution

First, we need to load the libraries (tools) we will use.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np
```

```python
path_url = 'https://raw.githubusercontent.com/industrial-data/predictor-explainer/main/data/dist_tower_na.csv'

#path_gclab = '/content/dist_tower_na.csv' # needs to be manually upladed first

df = pd.read_csv(path_url, index_col='Date', parse_dates=True)

# Parsing dates can be tricky, here we load the date as an indexed column.
```

```python
df.head() # First five rows
```

```python
df = df.reset_index() # (optional) add date as additional column (instead of index)
```

```python
df.plot( x='Date', y='yield', style='.-')
```

```python
df.describe()
```

# Data preparation

```python
df.info()
```

```python
# Data contains missings that will break the regression model if not filled

df = df.fillna(method="ffill") # forward filling (we use last known value until a new one arrives)
```

```python
target_colname = "yield"

y = df[target_colname].copy()
```

```python
X = df.copy()
```

```python
# Removing date and yield from model inputs
del X['Date']
del X[target_colname]
```

```python
# Adding synthetic noise as a model inpput to ease variable variable selection 
X["random_noise"] =  np.random.normal(0,1,len(X))
```

# Modeling

```python
forest = RandomForestRegressor()
```

```python
forest.fit(X,y)
```

```python
importances = forest.feature_importances_
```

```python
df_model_importances = pd.DataFrame(data={"Variable importance": forest.feature_importances_}, index=list(X.columns))

df_model_importances_sorted = df_model_importances.sort_values(by=['Variable importance'])

df_model_importances_sorted.plot.barh(y="Variable importance", figsize=(6,10))
```

```python
df.plot.scatter(x="Temp1", y="FlowC1",  c='yield',  cmap='viridis')
```

```python
df.plot.hexbin(x="Temp1", y="FlowC1",  C='yield', reduce_C_function=np.max, gridsize=15, cmap='viridis');
```

# Feature engineering (second model)


Generating additional features as model inputs

```python
for column in X:
  new_column_name =  "diff(" + str(column) + ")"
  try:
    X[new_column_name] = X[column].diff()
  except:
    print("Problems with", new_column_name)
```

```python
X.head()
```

```python
X = X.fillna(method="bfill")
```

```python
forest = RandomForestRegressor(random_state=42)
```

```python
forest.fit(X,y)
```

```python
importances = forest.feature_importances_
importances_std = np.std([
    tree.feature_importances_ for tree in forest.estimators_], axis=0)
```

```python
df_model_importances = pd.DataFrame(data={"Variable importance": forest.feature_importances_, "std(Importance)": importances_std},
                                     index=list(X.columns))

df_model_importances_sorted = df_model_importances.sort_values(by=['Variable importance'])

df_model_importances_sorted.plot.barh(y="Variable importance", figsize=(6,10), yerr=importances_std)
```

```python
top3_predictors = list(df_model_importances.sort_values(by=['Variable importance'], ascending=False).index[0:3])
```

```python
print(top3_predictors)
```

```python
df_results = X[top3_predictors].copy()
```

```python
df_results[target_colname] = y.copy()
```

```python
df_results
```

```python
for column in df_results.columns:
    max_value = df_results[column].max()
    min_value = df_results[column].min()
    if max_value != min_value:
        new_column_name =  str(column) + "_norm"
        df_results[new_column_name] = (df_results[column] - min_value) / (max_value - min_value)
```

```python
df_results['yield_cat'] = pd.cut(df_results[target_colname], bins=3, labels=['low', 'medium', 'high'])
```

```python
df_results[['yield', 'yield_cat']].head()
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10,6))

parallel_coordinates(df_results[["yield_norm","Temp1_norm", "diff(PressureC1)_norm","FlowC1_norm","yield_cat"]], "yield_cat")
```
